```python
import pandas as pd

def extract_pokemon_data():
    pokemon_data = pd.read_csv('pokemon.csv', index_col='#')
    return pokemon_data
```

```python
def clean_pokemon_data(pokemon_data):
    #Convert Name and Type columns from objects to strings
    pokemon_data['Name'] = pokemon_data['Name'].astype(str)
    pokemon_data['Type 1'] = pokemon_data['Type 1'].astype(str)
    pokemon_data['Type 2'] = pokemon_data['Type 2'].astype(str)

    # Replace a Nan for a missing type with None
    pokemon_data['Type 2'] = pokemon_data['Type 2'].replace('nan', 'None')
    pokemon_data['Type 1'] = pokemon_data['Type 1'].replace('nan', 'None')
    return pokemon_data

def relevant_pokemon_data(pokemon_data):
    return pokemon_data.drop(columns=['Name', 'Generation', 'Legendary'])
```

```python
def extract_clean_relevant_pokemon_data():
    clean_data = clean_pokemon_data(extract_pokemon_data())
    return relevant_pokemon_data(clean_data)
```
